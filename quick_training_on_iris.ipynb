```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# for training
data = pd.read_excel("iris_data_train.xlsx");
```

```python
data.head()
```

```python
x = data.iloc[:, :-1]
```

```python
y = data.iloc[:, -1]
```

```python
y
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x.values, y.values)
```

```python
predictions_1 = model.predict([[
    6,2.2,5,1.5
]])

predictions_2 = model.predict([[6.3,2.5,4.9,1.5]])
```

```python
predictions_1
```

```python
predictions_2
```

```python
datatest = pd.read_excel("iris_data_test.xlsx");
```

```python
x_test = datatest.iloc[:, :-1]
```

```python
predictions_test = model.predict(x_test.values)
```
